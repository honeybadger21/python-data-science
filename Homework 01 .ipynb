#### Prompt 01: 
The file FloridaVoters.html contains a Web Table of republican and democratic voters in various counties in Florida. Write code that reads in this file as a standard text file and prints out the counties, along with the number of republican and democratic voters in those counties, sorted by the number of democratic voters. 

```python
import numpy as np
import pandas as pd
import re

fp = open('FloridaVoters.html', 'r')

lines1 = fp.read() 
lines = lines1.replace(",","")

matches_body = re.findall('<td>\w+', lines,re.MULTILINE)

array_items = np.array(matches_body).reshape(-1, 3).tolist()

for index, value in enumerate(array_items):
    if index/2!=0:
        del array_items[index]
        
replaced_array = np.char.replace(array_items, '<td>', ' ', count = None)
columnIndex = 2

sortedArr = replaced_array[replaced_array[:,columnIndex].argsort()]

df = pd.DataFrame(sortedArr, columns = ['County','Republician','Democrat'])
df['Democrat'] = df['Democrat'].astype(int)

df = df.sort_values(by=["Democrat"])
df = df.reset_index(drop=True)

df
```

Task 1: Convert each tweet into a dictionary of phrases: Write a function called moving window that takes as input a tweet (that is, a string), and converts it into a bunch of phrases. Each phrase is three consecutive words in the tweet. For example, the tweet ‘‘This is an awesome tweet, dude’’ consists of the phrases ‘‘ This is an’’, ‘‘is an awesome’’, ‘‘an awesome tweet,’’, and ‘‘awesome tweet, dude’’. Return a dictionary whose keys are these phrases, and just set the corresponding values to 1. This dictionary contains all the unique 3-word phrases in the tweet.

```python
import pandas as pd
tweet = open('Santa.txt')

line = tweet.readline()
tweet_str = line

for line in tweet:
    tweet_str+=line

tweet_str = tweet_str.split("\n")
tweet_str.pop()

def moving_tweet(string_tweet):
    
    all_phrases = string_tweet.split(" ")
    
    k = len(all_phrases)
    i = 0
    
    DoP = {}
    
    for i in range(1, k-2):
        phrases = all_phrases[i-1] + " " + all_phrases[i] + " " + all_phrases[i+1]
        if phrases not in DoP:
            DoP[phrases] = 1        
        
    return DoP

for elem in tweet_str:
    print(moving_tweet(elem))
```

Write a function called cosine that takes as input two dictionaries. Each dictionary contains the 3-word phrases from one tweet. Return the cosine similarity between the phrases in the two dictionaries.

```python
def cosine(d1, d2):
    
    d1_new, d2_new = [], []
    
    for elem in d1:
        d1_new.append(elem)
        
    for elem in d2:
        d2_new.append(elem)
        
    if len(d1_new) == 0 or len(d2_new) == 0:
        return 0
    
    match = []
    
    for elem in d1_new:
        if elem in d2_new and elem not in match:
            match.append(elem)
    
    return len(match)/((len(d1_new)*len(d2_new))**0.5)

# Sample Function Call

d1 = {'SPECIAL SECRET HEARTS:': 1, 'SECRET HEARTS: A': 1, "HEARTS: A Child's": 1, "A Child's Introduction": 1, "Child's Introduction to": 1, 'Introduction to Dementia': 1, 'to Dementia and': 1, 'Dementia and Pink': 1, 'and Pink Curls': 1, 'Pink Curls -': 1, 'Curls - A': 1, '- A Santa': 1, 'A Santa ...': 1, 'Santa ... -': 1, '... - http://t.co/UWCdc8FA9a': 1}
d2 = {'"Santa Claus Is': 1, 'Claus Is Coming': 1, 'Is Coming To': 1, 'Coming To Town': 1, 'To Town #MTVHottest': 1, 'Town #MTVHottest Justin': 1}
d3 = {'RT @BuyBookstore: SPECIAL': 1, '@BuyBookstore: SPECIAL SECRET': 1, 'SPECIAL SECRET HEARTS:': 1, 'SECRET HEARTS: A': 1, "HEARTS: A Child's": 1, "A Child's Introduction": 1, "Child's Introduction to": 1, 'Introduction to Dementia': 1, 'to Dementia and': 1, 'Dementia and Pink': 1, 'and Pink Curls': 1, 'Pink Curls -': 1, 'Curls - A': 1, '- A Santa': 1, 'A Santa ...': 1, 'Santa ... -': 1, '... - http://t.co/UWCdc8FA9a': 1}

print("Cosine Similarity for d1 & d2:", cosine(d1, d2))
print("Cosine Similarity for d1 & d3:", cosine(d1, d3))
```

#### Prompt 03:
The file “quotes.txt” contains pairs of lines, with the first line being a quote and the following line being the person who said it. We want to build a search engine that, given a word or words, finds the best matching quotes.

Task 1: Build a list of full quotes: Read in the file, and create a list of full quotes of the form “quote - speaker”. For example, “The heart has its reasons, of which the mind knows nothing. - Blaise Pascal”.

```python
# Import Libraries

import re
import math
quotes_file = open("quotes.txt", "r")
quotes_file_to_string=quotes_file.read()

split_quotes= quotes_file_to_string.split('\n')

required_list=[]

for idx,i in enumerate(split_quotes):
    if idx==0:
        req=i
    elif idx < len(split_quotes)-1:
        if i[-1]=='.':
            required_list.append(req)
            req=i
        else:
            req=req+" -"+i
required_list.append(req)

print(required_list) 
```

Task 2: Words from full quotes: Write a function that takes a full quote as argument and outputs a list of the words in the it. The words should all be lower-case, and should contain only characters, digits, or underscore. Hint: Use the lower() function of strings, and re.split() to split into words, but you must figure out the regular expression for re.split().

```python
def function_split_quote(string):
    string_clean=string.lower()
    string_clean_2=re.sub('^[\W+_]', '', string_clean)
    word_list = re.split(r"\s+", string_clean_2.replace('-',''))
    print(word_list)
    return word_list
 
function_split_quote(required_list[0])
```

Task 3: Build the postings-list dictionary: A postings-list is a dictionary whose keys are full quotes, and whose values are themselves dictionaries with key being a word, and value being the number of times the word occurs in the full quote. So, for the key “The heart has its reasons, of which the mind knows nothing. - Blaise Pascal”, the value will itself be the following dictionary: {’the’:2, ’heart’:1, ’has’:1, ’its’:1, ’reasons’:1, ’of’:1, ’which’:1, ’mind’:1, ’knows’:1, ’nothing’:1,’blaise’:1, ’pascal’:1}.

```python
posting_list={}
def postings_list(input_string):
    dictionary={}
    clean_string=input_string.replace('-','').lower()
    words = clean_string.split()

    for idx, word in enumerate(words):
        if idx < len(words):
            phrase = word
            if phrase in dictionary:
                dictionary[phrase] += 1
            else:
                dictionary[phrase] = 1
    #print(dictionary)
    posting_list[input_string]=dictionary
    return posting_list

for item in required_list:
    postings_list(item)
    
print(posting_list)
```

Task 4: Build the reverse postings-list dictionary: A reverse postings-list is a dictionary whose keys are the words, and the values are themselves dictionaries with the key being a full quote, and the value being the number of times the word appeared in the full quote. So, for the key “entertainer”, the value is the dictionary {’An actor is at most a poet and at least an entertainer. - Marlon Brando’: 1} (only this quote contains the word “entertainer”).

```python
reverse_posting_list={}

for key in posting_list:
    value=posting_list[key]
    for key2 in value:
        reverse_posting_list[key2]={key:value[key2]}

print(reverse_posting_list)
```

Task 5: Write a TF-IDF function (8 points). To measure how much a full quote is about a particular word, one typically uses the TF-IDF measure.
• TF stands for “term frequency”; the term frequency of a word w in a full quote q is the number of times w occurs in q, divided by the maximum number of times any word occurs in q.
• IDF stands for “inverse document frequency”: the IDF of a word w is the logarithm of the ratio of the total number N of full quotes to the number of full quotes in that contain the word w.
• TF-IDF of a word w for a full quote q is just the product of the TF and IDF. 

So, for the word “entertainer” in the Marlon Brando quote of part (d):
• The TF is 0.5 (it occurs once, while the most frequent word in that quote is “at”, which occurs twice, so the TF ratio is 0.5)
• The IDF is log (886/1), since there are 886 documents and the word “entertainer ” occurs in only one full quote. 

Write a function to compute the TF-IDF of any word in any full quote, using the postings and reverse-postings.

```python
def tf_idf_calculator(word,quote):
    #calculate tf
    posting=posting_list[quote]
    #print(posting)
    new_val = posting.values()
    maximum_val = max(new_val)
    Frequency_of_word=posting[word]
    #print(Frequency_of_word,maximum_val)
    tf=Frequency_of_word/maximum_val
    #print(tf)
    
    #calculate idf
    total_quotes=len(posting_list)
    counter=0
    for key in posting_list:
        value=posting_list[key].keys()
        #print(value)
        if word in value:
            #string_ct=str(value).count(word)
            counter=counter+1
        #print(string_ct)
    #print(counter)
    
    idf=math.log(total_quotes/counter)
    #print(idf)
    
    tf_idf=idf*tf
    #print(tf_idf)
    
    return tf_idf

tf_idf_calculator('how','How we spend our days is, of course, how we spend our lives. -Annie Dillard')
```

Task 6: Quote search using a single word (5 points). Write a function that takes a word as argument, and returns a dictionary whose keys are full quotes containing that word, and whose values are the TF-IDF score of that
word for that full quote.

```python
dictionary_searched_single={}

def quote_search(word):
    for key in posting_list:
        value=posting_list[key].keys()
        #print(key,value)
        if word in value:
            dictionary_searched_single[key]=tf_idf_calculator(word,key)
    return dictionary_searched_single
        
quote_search('how')
dictionary_searched_single
```
